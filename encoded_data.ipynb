```python

# initial imports
# Load data
import pandas as pd
from path import Path
file_path = Path("Resources/loans_data.csv")
loans_df = pd.read_csv(file_path)
loans_df.head()
```

```python
# Months dictionary
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
```

```python
loans_binary_encoded = pd.get_dummies(loans_df, columns=["education", "gender"])
loans_binary_encoded['month_nums'] = loans_df['month'].apply(lambda x : months_num[x])
encoded_df = loans_binary_encoded.drop(['month'] ,axis=1)

encoded_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = loans_df.copy()
df2['education'] = le.fit_transform(df2['education']) 
df2['gender'] = le.fit_transform(df2['gender']) 
df2["month_num"] =df2["month"].apply(lambda x : months_num[x])

df2.head()
```
